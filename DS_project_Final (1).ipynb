```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Assuming your DataFrame is named 'df' and has columns 'Patients_Admitted' and 'Deaths'
# Replace with the actual column names if they are different

# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(df['Patients_Admitted'], df['Deaths']) # Changed 'Patient Admitted' to 'Patients_Admitted'

# Create the scatter plot with regression line using seaborn
sns.regplot(x='Patients_Admitted', y='Deaths', data=df, scatter_kws={'s': 20}, line_kws={'color': 'red'}) # Changed 'Patient Admitted' to 'Patients_Admitted'


# Add title and labels
plt.title(f'Scatter Plot with Linear Regression\nPearson Correlation: {correlation_coefficient:.2f}, p-value: {p_value:.3f}')
plt.xlabel('Patients_Admitted') # Changed 'Patient Admitted' to 'Patients_Admitted'
plt.ylabel('Deaths')

# Show the plot
plt.show()
```
